```python
sql(
    '''
    SELECT  order_id, 
            date_format(order_date, '%Y-%m-%d') order_date, 
            category,
            Avg(profit) OVER (PARTITION BY Order_Date) daily_avg_profit
    FROM superstore
    WHERE order_date BETWEEN (SELECT MIN(order_date) FROM superstore) AND 
                            (SELECT MAX(order_date) FROM superstore)
    GROUP BY order_date;
    '''
)
```
