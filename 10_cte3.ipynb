```python
sql(
    '''
    WITH daily_avg_sales AS
        (SELECT
            date_format(order_date, '%Y-%m-%d') order_day,
            avg(sales) avg_sales
        FROM 
            superstore
        GROUP BY
            order_day
        )
    SELECT order_day, avg_sales
    FROM daily_avg_sales
    ORDER BY order_day;
    '''
)
```
